## IMPORTING LIBRARIES

```python
# Data manipulation and handling
import pandas as pd

# Google Play Store Scraper
from google_play_scraper import app, reviews, Sort  # For scraping app details and reviews from Google Play Store

# Apple App Store Scraper
from app_store_scraper import AppStore  # For scraping app details and reviews from Apple App Store
```

# GOOGLE PLAY STORE - APP REVIEW DATA GATHERING

## LIST OF ALL APPS

```python
# List of App Names
all_apps = [
    'com.whatsapp', # WhatsApp Messenger
    'co.uk.getmondo', # Monzo - Mobile Banking
    'bbc.mobile.news.ww', # BBC News
    'com.mobispector.bustimes', # Bus Times - Live Public Transit
    'com.northcube.sleepcycle' # Sleep Cycle: Sleep Tracker
]
```

## FETCHING USER REVIEWS OF ALL APPS

```python
# Initialize an empty DataFrame to store all reviews
google_reviews_df = pd.DataFrame()
```

```python
# Function to fetch reviews for a single app
def fetch_reviews(app_id):
    # Extract the app name
    app_details = app(app_id)
    app_name = app_details['title']

    # Extract reviews
    result, _ = reviews(
        app_id,
        lang='en', # Language code for the reviews
        country='gb', # Country code for the UK
        sort=Sort.MOST_RELEVANT, # Sort reviews by most relevant
        count=5000 # Number of reviews to extract
    )

    # Convert to DataFrame and add app name
    reviews_df = pd.DataFrame(result)
    reviews_df['source'] = app_name
    return reviews_df

# Iterate over each app in the app_info list and collect reviews
for app_id in all_apps:
    print(f"Fetching reviews for {app_id}...")
    app_reviews_df = fetch_reviews(app_id)
    google_reviews_df = pd.concat([google_reviews_df, app_reviews_df], ignore_index=True)
```

## EXPORTING ALL APP REVIEWS

```python
# Save all reviews to a CSV file
google_reviews_df.to_csv('/content/drive/MyDrive/Dissertation24/Google_Play_Reviews.csv', index=False)
print("Reviews have been saved to Google_Play_Reviews.csv")
```

# APPLE APP STORE - APP REVIEW DATA GATHERING

## LIST OF ALL APPS

```python
# List of App Names
app_info = [
    ('389801252', 'instagram'), # Instagram
    ('1001501844', 'deliveroo-food-shopping'), # Deliveroo: Food & Shopping
    ('932493382', 'revolut-mobile-finance'), # Revolut - Mobile Finance
    ('469463298', 'citymapper-your-transport-app'), # Citymapper: Your Transport App
    ('493145008', 'headspace-meditation-sleep') # Headspace: Meditation & Sleep
]
```

## FETCHING USER REVIEWS OF ALL APPS

```python
# Initialize an empty DataFrame to store all reviews
apple_reviews_df = pd.DataFrame()
```

```python
# Function to fetch reviews for a single app
def fetch_reviews(app_id, app_name):
    app_store_app = AppStore(country="gb", app_name=app_name, app_id=app_id)

    # Extract reviews
    app_store_app.review(how_many=5000)

    # Convert to DataFrame and add app name
    reviews_df = pd.DataFrame(app_store_app.reviews)
    reviews_df['source'] = app_name
    return reviews_df

# Iterate over each app in the app_info list and collect reviews
for app_id, app_name in app_info:
    print(f"Fetching reviews for {app_name}...")
    app_reviews_df = fetch_reviews(app_id, app_name)
    apple_reviews_df = pd.concat([apple_reviews_df, app_reviews_df], ignore_index=True)
```

## EXPORTING ALL APP REVIEWS

```python
# Save all reviews to a CSV file
apple_reviews_df.to_csv('/content/drive/MyDrive/Dissertation24/App_Store_Reviews2.csv', index=False)
print("Reviews have been saved to App_Store_Reviews2.csv")
```
